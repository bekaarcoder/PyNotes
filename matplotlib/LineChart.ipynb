```python
from matplotlib import pyplot as plt
```

```python
print(plt.style.available)
plt.style.use('seaborn-dark')
```

```python
dev_x = [x for x in range(25, 36)]
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62381, 64936, 62846, 63846, 72673]
py_dev_y = [39496, 45000,50752, 49920, 56200, 59000, 66381, 65936, 69546, 68046, 76873]
```

```python
plt.plot(dev_x, dev_y, color='#c1b2ab', linestyle='--', marker='.', label='All Devs')
plt.plot(dev_x, py_dev_y, color='#4a5899', marker='.', linewidth=2, label='Python Devs')

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

plt.legend()

plt.grid(True)
plt.tight_layout()
```

```python
plt.style.use('fivethirtyeight')

plt.plot(dev_x, dev_y, linestyle='--', label='All Devs')
plt.plot(dev_x, py_dev_y, label='Python Devs')

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

plt.legend()

plt.tight_layout()
```
