```python
slices = [60, 40, 30, 20]
labels = ['Sixty', 'Forty', 'Thirty', 'Twenty']
colors = ['#5aa9e6', '#ff6392', '#2fbf71', '#ffe45e']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'white'}, colors=colors)

plt.title("Pie Chart")
plt.tight_layout()
```

```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('data/languages_data.csv')
ids = df['Respondent']
lang_responses = df['LanguageWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(";"))

labels = []
slices = []

for item in language_counter.most_common(5):
    labels.append(item[0])
    slices.append(item[1])

print(labels, slices)
```

```python
explode = [0, 0, 0, 0.1, 0]
# explode is used to put an emphasis on a pie
# startangle rotates the pie chart with the provided value
```

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'white'}, explode=explode, startangle=90, autopct='%1.1f%%')

plt.title("Popular Programming Languages")
plt.tight_layout()
```
